```python
import tensorflow as tf
```

```python
height = [170, 180, 175, 160]
size = [260, 270, 265, 255]
```

```python
a = tf.Variable(0.1)
b = tf.Variable(0.2)

def 손실함수():
#     손실값 = tf.square(실제값 - 예측값)
    예측값 = height * a + b
    손실값 = tf.square(size - 예측값)
    return 손실값

opt = tf.keras.optimizers.Adam(learning_rate = 0.1) # 러닝 레이트 기본값 0.001?

for _ in range(300):
    opt.minimize(손실함수, var_list = [a, b])
    print(a.numpy(), b.numpy())
    
```

```python
# 데이터 준비
train_x = [1,2,3,4,5,6,7]
train_y = [3,5,7,9,11,13,15]
```

```python
# 모덴 만들기

a = tf.Variable(0.1)
b = tf.Variable(0.1)

def 손실함수():
    예측_y = train_x * a + b
    return tf.keras.losses.mse(train_y, 예측_y)

opt = tf.keras.optimizers.Adam(learning_rate = 0.1)

for _ in range(1000):
    opt.minimize(손실함수, var_list = [a, b])
    print(a.numpy(), b.numpy())
```
